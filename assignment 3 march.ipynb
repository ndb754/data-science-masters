```python
df=sns.load_dataset('titanic')
# Create the box plot for age
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="pclass", y="age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Box Plot of Age by Passenger Class")
plt.show()

# Create the box plot for fare
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="pclass", y="fare")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Box Plot of Fare by Passenger Class")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset using seaborn
df = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut' column
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="price", hue="cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Price with Cut")
plt.show()
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
import seaborn as sns

# Load the "flights" dataset using seaborn
df = sns.load_dataset("flights")

# Reshape the data to create a pivot table
df_pivot = df.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(data=df_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Flights")
plt.show()
```

```python
'''In this example, we first use pivot to reshape the data into a pivot table where the rows represent the months,
the columns represent the years, and the values represent the number of passengers. 
Then, we use sns.heatmap to create the heatmap, with the data parameter set to the pivot table.
We also set annot=True to display the values in each cell, fmt="d" to format the values as integers, and cmap="YlGnBu" to set the color map.

Finally, we add labels and a title to the plot using plt.xlabel, plt.ylabel, and plt.title.'''
```
