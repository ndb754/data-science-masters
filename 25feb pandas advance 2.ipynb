```python
data1=pd.DataFrame({'key1':[1,2,3,4],'key2':[4,5,6,7],'key3':[2,3,4,5]})
```

```python
data2=pd.DataFrame({'key1':[1,2,5,4],'key4':[14,15,16,17],'key5':[21,31,41,51]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2) #merge function requires 1 common column between 2 table with common values 
```

```python
pd.merge(data1,data2,how='left') #will extract everything from data1 and only common values of column key1 from data2 table
```

```python
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='outer') # by default merge function is 'inner' ie only common values data it exctract
```

```python
pd.merge(data1,data2,how='cross')
```

```python
pd.merge(data1,data2,how='left',on='key1') # on is also used to selection common column
```

```python
pd.merge(data1,data2,how='left',left_on='key1',right_on='key4') #matching key1 of data1 with key4 of data2
```

```python
# now join()...merge and join are different type of functions
data3=pd.DataFrame({'key1':[1,2,3,4],
                    'key2':[4,5,6,7],
                    'key3':[2,3,4,5]},
                  index=['a','b','c','d'])
```

```python
data4=pd.DataFrame({'key1':[1,2,3,4],
                    'key4':[14,15,16,17],
                    'key5':[21,31,41,51]},
                   index=['a','b','d','e'])
```

```python
data3
```

```python
data4
```

```python
data3.join(data4) # join() does join on basis of index and not column values whereas merge does the join on basis of common values of common columns...it shouldnt show error but showing
```

```python
data3.join(data4,how='inner')
data3.join(data4,how='right')
```
