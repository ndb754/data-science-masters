```python
'''Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
In pandas, the term "window function" refers to a class of functions that operate on a set of rows (a window) of a DataFrame and return a new DataFrame with the same number of rows but with the values aggregated or transformed according to some specified rule.
Some common use cases for window functions include:

Moving averages and moving sums

Rolling statistics: These window functions calculate statistical measures such as mean, standard deviation, variance, or correlation for a fixed-size window of rows.

Expanding windows: These window functions calculate cumulative statistics such as the cumulative sum or mean up to a given row.

Shifts: These window functions shift the values of a column forward or backward by a certain number of rows.

Some of the commonly used window functions in pandas include:

rolling(): This function creates a rolling window of a fixed size and applies a function to each window. The window can be centered or left-aligned or right-aligned, and it can slide by a fixed or variable number of rows.

expanding(): This function creates an expanding window that starts from the first row and includes all previous rows up to the current row. It applies a function to each window.

ewm(): This function creates an exponentially weighted moving window of a fixed size and applies a function to each window.

shift(): This function shifts the values of a column forward or backward by a certain number of rows.

rank(): This function assigns ranks to each row in a window based on a specified criteria.

cumsum(), cumprod(), cummin(), cummax(): These functions calculate cumulative sums, products, minimums, or maximums over a window.

diff(): This function calculates the difference between the values of a column in consecutive rows.

pct_change(): This function calculates the percentage change between the values of a column in consecutive rows.'''
```

```python
'''Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.'''

import pandas as pd

# Prompt the user to enter the first date in string
date1_str = input("Enter the first date (YYYY-MM-DD): ") # converting string to date_time
date1 = pd.to_datetime(date1_str)

# Prompt the user to enter the second date in string
date2_str = input("Enter the second date (YYYY-MM-DD): ")
date2 = pd.to_datetime(date2_str)

# Calculate the time delta between the two dates
delta = date2 - date1

# Calculate the number of days, hours, and minutes in the time delta
days = delta.days
hours = delta.seconds // 3600
minutes = (delta.seconds // 60) % 60

# Display the result
print("The difference between", date1.date(), "and", date2.date(), "is:")
print(days, "days,", hours, "hours, and", minutes, "minutes.")
```

```python
'''Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.'''
import pandas as pd

# Prompt the user to enter the file path
file_path = input("Enter the path to the CSV file: ")

# Prompt the user to enter the column name
column_name = input("Enter the name of the column to convert: ")

# Prompt the user to enter the category order
category_order_str = input("Enter the category order (comma-separated): ")
category_order = [cat.strip() for cat in category_order_str.split(",")]

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Convert the specified column to a categorical data type
df[column_name] = pd.Categorical(df[column_name], categories=category_order)

# Sort the DataFrame by the specified column
df_sorted = df.sort_values(column_name)

# Display the sorted data
print(df_sorted)
```

```python
'''Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.'''


import pandas as pd
import matplotlib.pyplot as plt

# Prompt the user to enter the file path
file_path = input("Enter the path to the CSV file: ")

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Set the date column as the index and convert it to a datetime object
df = df.set_index('date')
df.index = pd.to_datetime(df.index)

# Group the DataFrame by product category and resample it by month, summing the sales data
df_monthly = df.groupby('product_category').resample('M').sum()

# Create a stacked bar chart showing the sales of each product category over time
ax = df_monthly.unstack('product_category').plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the chart title and axes labels
ax.set_title('Monthly Sales by Product Category')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')

# Display the chart
plt.show()
```

```python
'''Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
The program should do the followingM
I Prompt the user to enter the file path of the CSV file containing the student dataR
I Read the CSV file into a Pandas DataFrameR
I Calculate the mean, median, and mode of the test scores using Pandas toolsR
I Display the mean, median, and mode in a table.
Assume the CSV file contains the following columnsM
I Student ID: The ID of the studentR
I Test Score: The score of the student's test.
Example usage of the program:
Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value |
+-----------+--------+
| Mean | 79.6 |
| Median | 82 |
| Mode | 85, 90 |
+-----------+--------+
Assume that the CSV file student_data.csv contains the following data:
Student ID,Test Score
1,85
2,90
3,80
4,75
5,85
6,82
7,78
8,85
9,90
10,85'''# Prompt user to enter file path
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Read CSV file into Pandas DataFrame
df = pd.read_csv(file_path)

# Calculate mean, median, and mode of test scores
mean_score = df["Test Score"].mean()
median_score = df["Test Score"].median()
mode_score = df["Test Score"].mode()

# Display results in a table
result_table = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode"],
    "Value": [mean_score, median_score, mode_score]
})
print(result_table)
```
