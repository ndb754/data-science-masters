```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='b',alpha=0.8) # alpha is for intensity of color...it makes dark or light
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.zlabel('z axis')
```
